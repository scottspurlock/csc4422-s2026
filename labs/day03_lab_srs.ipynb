### Make one histogram for max height and another for max weight. Tune the number of bins so you can see how the data is distributed. Use a non-default color for each. (Try one of these color names.) Label the axes.

```python
plt.hist(df['max_height'], bins=12, color='darkorchid', edgecolor='k')
plt.xlabel('max_height')
plt.ylabel('frequency')
plt.show()
plt.hist(df['max_weight'], bins=12, color='slateblue', edgecolor='k')
plt.xlabel('max_weight')
plt.ylabel('frequency')
plt.show()
```

## Step 7
Display a table to answer the question: among small dogs (less than 10 pounds in terms of max weight), which group has the most different breeds? (try df.info() and look for a group column.)


```python
# small dogs
mask = df['max_weight'] < 10
small = df[mask]

# v1
# small.groupby('group')['breed'].count().sort_values(ascending=False)

# v2
small['group'].value_counts().plot(kind='barh', color='lightgreen')
plt.xlabel('Number of breeds')
plt.show()
```

## Step 9
Make up a question that requires filtering the data and grouping it. Then display a bar graph to answer the question.

"What's the average shedding value for each grooming frequency category?"

```python
mask = df['max_weight'] > 15
large = df[mask]
large.groupby('grooming_frequency_category')['shedding_value'].mean().sort_values(ascending=False).plot(kind='barh', color='pink')
plt.xlabel('Average shedding value')
plt.ylabel('Grooming frequency category')
plt.show()
```
