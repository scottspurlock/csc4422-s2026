```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

## Read in the public data

```python
public_url = 'https://drive.google.com/uc?export=download&id=1pRAvxdUr9SU9H5puLK7xBg256sWCO9VE'

df = pd.read_csv(public_url)
df.head()
```

## Read in the competition data
Note that while the public data includes the target, target_upgrade, the competition data doesn't, since the goal of the competition is to predict this.

```python
competition_url = 'https://drive.google.com/uc?export=download&id=1WVKCy9bbqZjHpZ0cjaPm7_4m4LoRobmd'

df_competition = pd.read_csv(competition_url)
df_competition.head()
```

## Next would be EDA and data wrangling
For simplicity, I'll just select a few numeric columns that are already ok for making a model. You'll want to do more to wrangle the data into shape.

```python
X = df[['zip_code', 'est_annual_spend', 'internal_metric_alpha']]
y = df['target_upgrade']

X.head()
```

## Next we would want to perform experiments with the public data to find the best type of model and best hyperparameters
For simplicity, I'm just going to train a basic KNN model without experimenting.

For this project, we care most about the average f1 score, which you can calculate using f1_score, as shown below, or find it in the more general classification_report.

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)

model = KNeighborsClassifier(n_neighbors=1)
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)

acc = accuracy_score(ytest, ypred)
f1 = f1_score(ytest, ypred, average='macro')
print(f'Our model has accuracy {acc:.2f} and f1 score {f1:.2f}')

print()

print(classification_report(ytest, ypred))
```

## Now use your trained model to predict the competition data
Make sure to do the same things to the competition data as you do to the public data

Notes:
- use the same columns for the competition data that the trained model is expecting.
- if you cleaned (or normalized, etc.) a column in the public data, clean the column in the competition data in the same way.
- don't drop any of the rows in the competition data, since you need to make predictions for all the rows.
- we don't have target_upgrade for the competition data, so we can't evaluate how well the model works on it here. Instead we'll upload the predictions to the competition site, which does have the target_upgrade, and will tell us the f1 score.

```python
Xcompetition = df_competition[['zip_code', 'est_annual_spend', 'internal_metric_alpha']]
Xcompetition.head()
```

## Once the final model is trained, we can predict the competition data

```python
ypred = model.predict(Xcompetition)
print('First 10 predictions:', ypred[:10])
```

## For submission to the Kaggle competition, we need to create a csv file
The format will have a column for customer_id and one for the predictions, and one row for each competition example.

### Notes
- Kaggle is expecting 2000 predictions. Make sure not to drop any of the rows from the competition data.
- Kaggle is expecting predictions to be 0's or 1's
- Kaggle needs the customer_id's to match up predictions with the right answers, so we'll use the customer_id's from the competition dataframe.

```python
ids = df_competition['customer_id']
predictions = ypred

# make a dataframe to hold the competition entry
output = pd.DataFrame({'customer_id': ids, 'target_upgrade': predictions})

output
```

Save the predictions as a CSV file

```python
output.to_csv('submission.csv', index=False)
print("Kaggle submission was successfully saved...")
```
